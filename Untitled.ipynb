```python
import os


os.makedirs("EngAlphabet")

# Створюємо основну папку "UkrAlphabet"
os.makedirs("UkrAlphabet")

# Створюємо папки для кожної літери українського алфавіту
for char in range(ord('A'), ord('Z')+1):
    folder_name = chr(char)
    os.makedirs(f"EngAlphabet/{folder_name}", exist_ok=True)
    os.makedirs(f"EngAlphabet/{folder_name}/uppercase", exist_ok=True)
    os.makedirs(f"EngAlphabet/{folder_name}/lowercase", exist_ok=True)

# Створення підтек в папці UkrAlphabet для великих та малих літер
for char in range(ord('А'), ord('Я')+1):
    folder_name = chr(char)
    os.makedirs(f"UkrAlphabet/{folder_name}", exist_ok=True)
    os.makedirs(f"UkrAlphabet/{folder_name}/uppercase", exist_ok=True)
    os.makedirs(f"UkrAlphabet/{folder_name}/lowercase", exist_ok=True)

# Створюємо папку "Digits"
os.makedirs("Digits")

# Створюємо папки для кожної цифри
for digit in range(10):
    folder_name = str(digit)
    os.makedirs(f"Digits/{folder_name}")
# Створюємо папку "Spec"
os.makedirs("Spec")

# Список спецсимволів
special_chars = ['.', ',', ';', ':', '!', '?', '-', '_', '(', ')', '[', ']', '{', '}', '<', '>', '/', '\\', '|', '+', '=', '*', '&', '%', '@', '#', '$', '^', '~', '`', '"', '\'']

# Словник заміни неприпустимих символів
replacement_dict = {
    '.': 'dot',
    ',': 'comma',
    ';': 'semicolon',
    ':': 'colon',
    '!': 'exclamation',
    '?': 'question',
    '-': 'dash',
    '_': 'underscore',
    '(': 'left_parenthesis',
    ')': 'right_parenthesis',
    '[': 'left_bracket',
    ']': 'right_bracket',
    '{': 'left_curly_brace',
    '}': 'right_curly_brace',
    '<': 'less_than',
    '>': 'greater_than',
    '/': 'forward_slash',
    '\\': 'backward_slash',
    '|': 'pipe',
    '+': 'plus',
    '=': 'equal',
    '*': 'asterisk',
    '&': 'ampersand',
    '%': 'percent',
    '@': 'at',
    '#': 'hash',
    '$': 'dollar',
    '^': 'caret',
    '~': 'tilde',
    '`': 'backtick',
    '"': 'double_quote',
    '\'': 'single_quote'
}

# Створюємо папку для кожного спецсимволу
for char in special_chars:
    if char in replacement_dict:
        folder_name = replacement_dict[char]
    else:
        folder_name = char
    os.makedirs(f"Spec/{folder_name}")
```
